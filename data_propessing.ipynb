```python
import cv2
import os
import glob
from tqdm import tqdm # 顯示進度條
```

路徑設置

```python
train_dir="./training_image/"
test_dir="./testing_image/"
```

基本分析物件創建

```python
# 創建 CLAHE 物件 (參數可調整，但這是醫療影像的常用預設值)
# clipLimit=2.0 (限制對比度過度放大)
# tileSize=(8, 8) (定義局部區域的大小)
clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileSize )
```

前處理的(寫成方法)

```python
def propessing(root_data_dir :str ):
    for root, dirs, files in os.walk(root_data_dir):
        # 2. 篩選出所有影像檔案
        image_files = [f for f in files if f.endswith(('.png', '.jpg', '.jpeg'))]

        # 3. 如果當前資料夾有影像檔案，則進行處理
        if image_files:
            print(f"\n正在處理資料夾: {root} ({len(image_files)} 張影像)...")
        
            # 使用 tqdm 顯示進度條
            for file_name in tqdm(image_files):
                file_path = os.path.join(root, file_name)
            
                # 讀取影像 (以灰階模式讀取)
                img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

                if img is None:
                    # 跳過讀取失敗的檔案
                    continue

                # 應用 CLAHE
                clahe_img = clahe.apply(img)

                # 儲存處理後的影像 (原地覆蓋原檔案)
                cv2.imwrite(file_path, clahe_img)

    print("\n--- CLAHE 遞迴處理完成！所有影像已優化。---")
```

```python
propessing(train_dir)
propessing(test_dir)
```
